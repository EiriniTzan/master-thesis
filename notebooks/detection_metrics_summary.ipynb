## Final detection metrics summary

The following table reports the average classic detection metrics across the five datasets, for each detector.

We compute:

- MDR as the mean missed detection indicator
- FAR as the mean false alarm indicator
- MTD as the mean normalized detection delay
- MTFA as the mean normalized time between false alarms

Lower MDR, FAR and MTD values indicate better detection behavior.

Higher MTFA values indicate fewer unnecessary alarms.

The Page-Hinkley row corresponds to the paper's STUDD configuration.

The remaining rows correspond to the extended detector comparison, where the internal STUDD detector is replaced by another drift detector.

```python
detector_order = [
    "PageHinkley",
    "ADWIN",
    "HDDMAverage",
    "HDDMWeighted",
    "CUSUM",
]

final_detection_summary = (
    all_detection_metrics_df
    .groupby("detector")
    .agg(
        MDR=("missed_detection", "mean"),
        FAR=("has_false_alarm", "mean"),
        MTD=("normalized_detection_delay", "mean"),
        MTFA=("normalized_mtfa", "mean"),
        mean_num_alarms=("num_alarms", "mean"),
        mean_num_false_alarms=("num_false_alarms", "mean"),
        datasets=("dataset", "nunique"),
    )
    .reset_index()
)

final_detection_summary["detector"] = pd.Categorical(
    final_detection_summary["detector"],
    categories=detector_order,
    ordered=True,
)

final_detection_summary = (
    final_detection_summary
    .sort_values("detector")
    .reset_index(drop=True)
)

final_detection_summary_display = final_detection_summary.copy()

metric_columns = [
    "MDR",
    "FAR",
    "MTD",
    "MTFA",
    "mean_num_alarms",
    "mean_num_false_alarms",
]

final_detection_summary_display[metric_columns] = (
    final_detection_summary_display[metric_columns]
    .round(4)
)

final_detection_summary_display
```
